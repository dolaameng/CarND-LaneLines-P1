Below are some helper functions to help get you started. They should look familiar from the lesson!

```python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    you should call plt.imshow(gray, cmap='gray')"""
    # I THINK IT SHOULD USE `cv2.COLOR_RGB2GRAY` HERE 
#     return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def merge_lines(lines, W, H):
    """Merge samll pieces of lines into two borders
    Return two main lines, each is represented by a single start/end point pair
    """
    # find all segments
    segments = np.array([seg for line in lines for seg in line])
    
    # get slopes and lengths of segments
    slopes = np.array([(y2-y1)*1. / (x2-x1+1e-6) for (x1, y1, x2, y2) in segments])
    lengths = np.array([np.sqrt((x2-x1)**2 + (y2-y1)**2) for (x1, y1, x2, y2) in segments])
    
    # select segments with length larger than 80% - to remove noise
    selected = lengths >= np.percentile(lengths, 20)
    left_part = selected & (slopes >= 0.3)
    right_part= selected & (slopes <= -0.5)
    
    # find the averaged slope for left and right border
    left_slope = np.median(slopes[left_part])
    right_slope = np.median(slopes[right_part])
    
    # return borders as left_bottom/right_bottom to half height
    def _fit_line(x, y, slope):
        start = ( (H-y)/slope + x, H)
        end = ( (H/2+40-y)/slope + x, H/2+40)
        return start, end
    
    # pick a representative point on each lane border and fit a line through them
    # pick the pionts from the longest segment
    
#     lx1, ly1, lx2, ly2 = zip(*segments[left_part])
#     left_x, left_y = (np.median(lx1)+np.median(lx2))/2, (np.median(ly1)+np.median(ly2))/2
    lx1, ly1, lx2, ly2 = segments[left_part][np.argmax(lengths[left_part])]
    left_x, left_y = (lx1+lx2)/2, (ly1+ly2)/2

    left_start, left_end = _fit_line(left_x, left_y, left_slope)
    
#     rx1, ry1, rx2, ry2 = zip(*segments[right_part])
#     right_x, right_y = (np.median(rx1)+np.median(rx2))/2, (np.median(ry1)+np.median(ry2))/2
    rx1, ry1, rx2, ry2 = segments[right_part][np.argmax(lengths[right_part])]
    right_x, right_y = (rx1+rx2)/2, (ry1+ry2)/2
    right_start, right_end = _fit_line(right_x, right_y, right_slope)
    
    borders = [[left_start + left_end], [right_start + right_end]]
    borders = np.asarray(borders, dtype=np.int)
    return borders

def draw_lines(img, lines, color=[255, 0, 0], thickness=10):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    H, W = img.shape[:2]
    borders = merge_lines(lines, W, H)
    for line in borders:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, edges, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(edges, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros(img.shape, dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
```

## Lane Border Detection Algoritm


```python
def detect_lane_borders(img):
    """Detect left and right border of lanes in the center of image.
    Assumptions:
    - Lane is in the center of image
    - There are two borders (left and right) for the lane
    - The borders are solid or dashed lines/curves in bright white and yellow
    """
    # instead of using grayscale, convert RGB to HSV and use V channel
    # for brightness pattern
    gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)[:,:,2]    
    # smooth it for futher processing
    gray_img = gaussian_blur(gray_img, kernel_size=7)
    
    # mask 1 as brightness from HSV
    brigtness_mask = (gray_img >= gray_img.max() * 0.5)
    # mask 2 as ROI - triangle in the centeral bottom
    H, W = gray_img.shape
    roi_mask = np.zeros_like(gray_img)
    vertices = np.array([[(0,H),(W/2, H/2+40), (W,H)]], dtype=np.int32)
    cv2.fillPoly(roi_mask, vertices, 255)
    # combined mask
    mask = np.bitwise_and(brigtness_mask, roi_mask)
    
    # masked canny detection
    edges = cv2.Canny(gray_img, threshold1=50, threshold2=150)
    edges = np.bitwise_and(edges, mask)
    
    # do a closing on edges
    # edges = cv2.morphologyEx(edges, cv2.MORPH_CLOSE, kernel=(50, 50))
    
    # hough transformations
    lines = hough_lines(img, edges, rho=1, theta=np.pi/180, threshold=1, min_line_len=10, max_line_gap=5)
    
    # combined orignal image with drawn lines
    result = weighted_img(lines, img, α=1, β=1)
    return result
```

## My Reflections on P1
There are two things that I see as improvement opportunities in my current implementation. 

- Part of the algorithm is based on certain assumptions and thus implemented in a "hard-coded" way. For example, I assume there are always a left and a right border in each image and they should always appear in the central locations. And the orientations of the borders should be roughly in a range.

  One disadvantage of hard coding these rules is its rigidity when dealing with new scenarios, e.g., with snows as white marks on the road in bad weathers? I believe that at least some of these hard-coded rules should be learned in a "data-driven" way, e.g., by seeing a lot of images and automatically learning to predict the locations of borders in an image.
  

- Compared to the provided solution in "P1_example.mp4", the lines detected in my current solution are a little bit "shaky" when moving from one video frame to the next. This indicates that the detection of lines is not smooth enough - small changes in the image could result in visual differences in the lines. 

  I suspect this is due to the "line merging" part of my code. I have tried to use robustic statistics such as "median of slopes" to ease it up. Improvements have been seen but there might be more to do in the future.
  
  
I have also tried the optional challenge below. The biggest challenge to my previous soluton is to correctly detect the borders when the colors of the lane changed at around 4th second.  

Restricting the slope ranges of lines to be considered as borders helps make my current implementation more robust to the color change. I also used "HSV" (rather than RGB) based grayscale image for edge detection, which helps distinguish bright borders from others such as the shadows of trees or other vehicles.
